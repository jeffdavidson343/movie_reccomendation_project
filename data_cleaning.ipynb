```python
movies_df = pd.read_csv(Path ("movies2.csv"))
movies_df
```

```python
genres_list = []
for index, row in movies_df.iterrows():
    try:
        genres = row.generes.split('|')
        genres_list.extend(genres)
    except:
        genres_list.append(row.generes)
        
genres_list = list(set(genres_list))
genres_list = [x for x in genres_list if str(x) != 'nan']
genres_list
```

```python
for index, row in movies_df.iterrows():

        for genre in genres_list:
                if genre in str(movies_df.loc[index, 'generes']):
                        movies_df.loc[index, genre] = 1
                else: 
                        movies_df.loc[index, genre] = 0
movies_df.to_csv("encoded_movies.csv")
movies_df
```

```python
users_df = pd.read_csv(Path ("users.csv"))

users_df = users_df.drop('zip_code', axis=1)
users_df
```

```python
AGES = { 1: "Under 18", 18: "18-24", 25: "25-34", 35: "35-44", 45: "45-49", 50: "50-55", 56: "56+" }
OCCUPATIONS = { 0: "other or not specified", 1: "academic/educator", 2: "artist", 3: "clerical/admin",
                4: "college/grad student", 5: "customer service", 6: "doctor/health care",
                7: "executive/managerial", 8: "farmer", 9: "homemaker", 10: "K-12 student", 11: "lawyer",
                12: "programmer", 13: "retired", 14: "sales/marketing", 15: "scientist", 16: "self-employed",
                17: "technician/engineer", 18: "tradesman/craftsman", 19: "unemployed", 20: "writer" }
```

```python
users_df["age_desc"]=users_df['age'].apply(lambda x: AGES[x])
users_df['occ_desc'] = users_df['occupation'].apply(lambda x: OCCUPATIONS[x])
users_df
```

```python
users=users_df[['userID','gender','zip_code', 'age_desc', "occ_desc"]]
users
```
